##**Objective 1:** To read the .xml data files with the crime reports from 2011, 2012 and 2013 and then print them out in form of dataframe.
    
* With the aim of extracting the crime incidents data from the xml data file I built up a function called "*dataframe_builder*"
that takes the file name as argument.

```python
def dataframe_builder(datafile):
    
    datafile=str(datafile)
    
    xmldoc = minidom.parse(datafile)
    itemlist = xmldoc.getElementsByTagName('dcst:ReportedCrime')
    
    OF=[]
    Wards=[]
    Dates=[]
    days=[]
    Streets=[]
    Shift=[]

    wards=['Ward1','Ward2','Ward3','Ward4','Ward5','Ward6','Ward7','Ward8']

    #Extraemos la info del xml siguendo el video de https://www.youtube.com/watch?v=c2qlCZhkwtE
    for s in itemlist[0:-1]:
        try:
            offenses=s.getElementsByTagName('dcst:offense')[0].firstChild.data
            ward=s.getElementsByTagName('dcst:ward')[0].firstChild.data 
            date=s.getElementsByTagName('dcst:reportdatetime')[0].firstChild.data
            street=s.getElementsByTagName('dcst:blocksiteaddress')[0].firstChild.data
            shift=s.getElementsByTagName('dcst:shift')[0].firstChild.data
            OF.append(offenses)
            Wards.append(ward)
            Dates.append(date)
            days.append(date.split('T')[0])
            Streets.append(street)
            Shift.append(shift)
        except:
            continue
    df = pd.DataFrame({'Ward' : Wards, 'Offenses':OF, 'Reported_Date':Dates, 'Street':Streets, 'Shifts':Shift, 'Days':days})
    return df
```

```python
df2013=dataframe_builder('crime_incidents_2013_plain.xml')
df2012=dataframe_builder('crime_incidents_2012_plain.xml')
df2011=dataframe_builder('crime_incidents_2011_plain.xml')

```

## Objective 4. Find the the probability of NOT being affected by any type of crime
Lets see what was the probability of NOT being affected by any type of crime
for a person who lived in DC the whole 2011, 2012 or 2013.
Define the random variable X: to be 100% safe in DC, and Y: to be attacked during 
any of the 365 days of any of these years at any location inside of DC

P(X) = 1 - P(Y)

P(Y) = ( number of outcomes / total number of possible outcomes )

in our case:

P(Y) = (number of crime reported incidents / total of DC inhabitants)

```python
# populations 
po2011=620427 #population_2011: http://www.census.gov/popest/estimates.html
po2012=635040 #population_2012: http://www.census.gov/popest/estimates.html
po2013=649111 #population_2011: http://www.census.gov/popest/estimates.html 

def prob_safety(df,population):
    PY= df.Offenses.value_counts().sum() / population
    # Then P(X)=1-P(Y)
    PX=round((1-PY)*100,2)
    return PX

safety_prob_2011=prob_safety(df2011,po2011)
safety_prob_2012=prob_safety(df2011,po2012)
safety_prob_2013=prob_safety(df2011,po2013)

print "The probabilities of NOT being affected by any type of crime in 2011, 2012 and 2013 were:",  safety_prob_2011,"%",safety_prob_2012,"%", safety_prob_2013,"%"        
```

##Objective 6. Find the probability per type of crime and check if they follow any trend
Lets see what are the probabilities per type of crime for a person who is living at any place in the city during 2011, 2012 or 2013.

```python
def P_offense_type(dataframe,population,year):
    '''calculates the probability that an inhabitant had during the given year 
    of being involved in the different reported crimes'''
    OF=[]
    Probs=[]
    year=str(year)
    for i,j in zip(dataframe.Offenses.value_counts().keys(),dataframe.Offenses.value_counts()):
       
        OF.append(i)
        Probs.append(round((j/population)*100,2))
    probs_crime_types_df=pd.DataFrame({'Offenses':OF, year+'_Probability(%)':Probs})
    return probs_crime_types_df



df_offense_type2011=P_offense_type(df2011,po2011,2011).set_index('Offenses')
df_offense_type2012=P_offense_type(df2012,po2012,2012).set_index('Offenses')
df_offense_type2013=P_offense_type(df2013,po2013,2013).set_index('Offenses')

dfs_per_crime=[df_offense_type2011,df_offense_type2012,df_offense_type2013]

df_percent_per_crime=pd.concat(dfs_per_crime, axis=1)
df_percent_per_crime
```

##Objective 8. Crime distribution per day's shift. Statistical analysis.

Let's see now which is the frequency of crimes within these three periodes of the day: day, evening and midnight.


**Hypothesis test for more than two proportions**

I would like to know if the proportions of crimes per shift changed with time. 
To do that, I'll use the **chi-square test**. The difference between the observed cell counts and those expected given no association (with fixed marginal totals) are compared to a table of the chi-square distribution. Large differences indicate discrepancies over and above chance and that there is a real association.

The following table compares the number of reported incidents per year among the three shifts.


```python
df_shifts=pd.DataFrame({'2011':df2011.Shifts.value_counts(),'2012':df2012.Shifts.value_counts(),
                        '2013':df2013.Shifts.value_counts()})

df_shifts
#plt.savefig("Figure9.png",dpi=300) # save as png
```

And the proportions are:

```python
def build_proportion_df():
    '''df with DAY, EVENING and MIDNIGHT proportions'''
    Shifts=['DAY', 'EVENING' , 'MIDNIGHT'] 
    proportions2011=[]
    proportions2012=[]
    proportions2013=[]
    years=['2011','2012','2013']
    for year in years:        
        for shift in Shifts:
            proportion=df_shifts[year].ix[shift]/df_shifts[year].sum()
            if year == '2011':
                proportions2011.append(round(proportion,4)*100)
            elif year == '2012':
                proportions2012.append(round(proportion,4)*100)
            else:
                proportions2013.append(round(proportion,4)*100)
            
            
    df=pd.DataFrame({'Shift':Shifts,'proportions2011':proportions2011,'proportions2012':proportions2012,'proportions2013':proportions2013})
    return df
prop_df=build_proportion_df()

prop_df.set_index('Shift')    
```

**There is a relation between the shift day and the crime ocurrence? **

Let's check it, here the hypothesis:
* H<sub>0</sub>: No relation 
* H<sub>A</sub>: There is a relationship


```python

DAY_total=round(df_shifts.ix['DAY'].sum(),2)
EVENING_total=round(df_shifts.ix['EVENING'].sum(),2)
MIDNIGHT_total=round(df_shifts.ix['MIDNIGHT'].sum(),2)


total_2011= df_shifts['2011'].sum()
total_2012= df_shifts['2012'].sum()
total_2013= df_shifts['2013'].sum()

TOTAL=DAY_total+EVENING_total+MIDNIGHT_total

#The proportion of crimes reported during the DAY is:
Day_proportion=round(DAY_total/TOTAL,4)
EVENING_proportion=round(EVENING_total/TOTAL,4)
MIDNIGHT_proportion=round(MIDNIGHT_total/TOTAL,4)

#Expected values (there are 9)
expected_CRIME_day_2011 = total_2011*Day_proportion
expected_CRIME_day_2012 = total_2012*Day_proportion
expected_CRIME_day_2013 = total_2013*Day_proportion

expected_CRIME_EVENING_2011 = total_2011*EVENING_proportion
expected_CRIME_EVENING_2012 = total_2012*EVENING_proportion
expected_CRIME_EVENING_2013 = total_2013*EVENING_proportion

expected_CRIME_MIDNIGHT_2011 = total_2011*MIDNIGHT_proportion
expected_CRIME_MIDNIGHT_2012 = total_2012*MIDNIGHT_proportion
expected_CRIME_MIDNIGHT_2013 = total_2013*MIDNIGHT_proportion

#print 'The overal proportion of crimes reported during the DAY is:',Day_proportion*100,'%'
```

```python
DA2011=((12853-expected_CRIME_day_2011)**2)/expected_CRIME_day_2011
DA2012=((14441-expected_CRIME_day_2012)**2)/expected_CRIME_day_2012
DA2013=((14094-expected_CRIME_day_2013)**2)/expected_CRIME_day_2013

EV2011=((14628-expected_CRIME_EVENING_2011)**2)/expected_CRIME_EVENING_2011
EV2012=((14686-expected_CRIME_EVENING_2012)**2)/expected_CRIME_EVENING_2012
EV2013=((15164-expected_CRIME_EVENING_2013)**2)/expected_CRIME_EVENING_2013

MI2011=((6481-expected_CRIME_MIDNIGHT_2011)**2)/expected_CRIME_MIDNIGHT_2011
MI2012=((6308-expected_CRIME_MIDNIGHT_2012)**2)/expected_CRIME_MIDNIGHT_2012
MI2013=((6638-expected_CRIME_MIDNIGHT_2013)**2)/expected_CRIME_MIDNIGHT_2013
```

```python
chi_2= DA2011+DA2012+DA2013+EV2011+EV2011+EV2011+MI2011+MI2012+MI2013
chi_2
```

```python
stats.chi2.pdf(chi_2 , 4)
```

The p-value is 1.4001126663860678e-14 indicating that we must reject the null hypothesis. Then, there is a relationship
among the shifts of the day and the crimes.

```python
months=['Jan','Feb','Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dic']

def crime_per_date_dataframe_builder(df,year):  
    
    df['Reported_Date'] = df['Reported_Date'].astype('datetime64[ns]')
    df_dates=df.set_index('Reported_Date')
    
    
    Jan = df[df['Reported_Date'].map(lambda x: x.month) == 1].Offenses.value_counts().sum()
    Feb = df[df['Reported_Date'].map(lambda x: x.month) == 2].Offenses.value_counts().sum()
    Mar = df[df['Reported_Date'].map(lambda x: x.month) == 3].Offenses.value_counts().sum()
    Apr = df[df['Reported_Date'].map(lambda x: x.month) == 4].Offenses.value_counts().sum()
    May = df[df['Reported_Date'].map(lambda x: x.month) == 5].Offenses.value_counts().sum()
    Jun = df[df['Reported_Date'].map(lambda x: x.month) == 6].Offenses.value_counts().sum()
    Jul = df[df['Reported_Date'].map(lambda x: x.month) == 7].Offenses.value_counts().sum()
    Aug = df[df['Reported_Date'].map(lambda x: x.month) == 8].Offenses.value_counts().sum()
    Sep = df[df['Reported_Date'].map(lambda x: x.month) == 9].Offenses.value_counts().sum()
    Oct = df[df['Reported_Date'].map(lambda x: x.month) == 10].Offenses.value_counts().sum()
    Nov = df[df['Reported_Date'].map(lambda x: x.month) == 11].Offenses.value_counts().sum()
    Dic = df[df['Reported_Date'].map(lambda x: x.month) == 12].Offenses.value_counts().sum()

    #months=['Jan','Feb','Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dic']
    incidents_month=[Jan,Feb,Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dic]
    month_df=pd.DataFrame({'Months':months, 'Number_of_incidents_'+str(year):incidents_month})
    
    return month_df.set_index('Months')

dates_dfs=[crime_per_date_dataframe_builder(df2011,2011), crime_per_date_dataframe_builder(df2012,2012), crime_per_date_dataframe_builder(df2013,2013)]

total=pd.concat(dates_dfs,axis=1)

total
```

Definitively here seems to be a possible **correlation between colder weather and criminal activity**. Lets check it.

To do that we need:
    * DF with the temperatures and the number of incidents per month for a given year
    * A plot to represent the linear regresion.
    
Then, first let's define a new function called Linear_regresion_df_builder that builts up the DF.

```python
from scipy.interpolate import *
from scipy.stats import *
from numpy import *

def Linear_regresion_df_builder(year):
    ''' Returns a DF with the temperatures and the number of incidents per month for a given year'''
    num_crime_year="Number_of_incidents_"+str(year)
    temp_year="T_C_"+str(year)
    T=[]
    num=[]
    for temperature,num_incidents in zip(tempe_df[temp_year],total[num_crime_year]):
        T.append(temperature)
        num.append(num_incidents)
    
    
    regresion_df= pd.DataFrame({'Temperatures_C':T,num_crime_year:num})
    
    return regresion_df
    
```

And now, define the plot builder function:

```python
def Plot_regresion(year):
    '''plot the data'''
    num_crime_year="Number_of_incidents_"+str(year)    
    regresion=Linear_regresion_df_builder(year)
    regresion.plot( kind='Scatter', x='Temperatures_C',y=num_crime_year,
                                   figsize=(6, 4), title='Num_Incidents VS Temperature_ '+str(year),
                                   fontsize=14)
    plt.plot()
    X=regresion.Temperatures_C
    Y=regresion[num_crime_year]  
    p1=polyfit(X,Y,1)
    plt.plot(X,polyval(p1,X),color='r')
    
    slope,intercept,r_value,p_value,std_err=linregress(X,Y)

    print "The r-square value is:", round((pow(r_value,2)),2),"and the p-value is:",round((pow(p_value,2)),4)
     
       
```

How does it look and how good is the adjustment (r square)?

```python
Plot_regresion(2011)
#plt.savefig("Figure8.png",dpi=300) # save as png
```

We can say then that the **36% of the variability** in the Number of incidents in 2011 with the temperatures **is explained by the model**.
Let's do **hypothesis testing** by seting the following hypothesis:

* H<sub>0</sub>= SLOPE of the regresion line = 0
* H<sub>A</sub>= SLOPE different than 0

As we got a p-value of 0.17% we can reject the null hypothesis. 

It means that we can say that there is a **very weak** possitive linear correlation between the number of incidents registered in 2011 and the temperatures.
